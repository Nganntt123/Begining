```python
# mỗi cột y cách nhau 30 đơn vị
from tkinter import *
import qrcode
from tkinter import filedialog
from tkinter import messagebox
from PIL import Image, ImageTk


window = Tk()
window.geometry('1000x550')
window.title("QR Code Generator")
# đổi màu nền bg
window.config(bg="#AE2321")
window.resizable(True, True)


# tạo hàm mở đường dẫn lưu file
direct = ""  # tạo biến lưu đường dẫn




def open_path():
   global direct
   files = [('PNG', '*.png')]
   direct = filedialog.asksaveasfilename(filetypes=files, defaultextension=files)
   entry_path.set(direct)
   if len(direct) == 0:
       msb_path = messagebox.showwarning("Cảnh báo", "Bạn hãy nhập đường dẫn lưu file")




# tạo hàm tạo mã qr
def getqr():
   data = (
       f"Họ và tên: {input.get()}\nNgày sinh: {input2.get()}\nMã số SV: {input3.get()}\nNgành học: {input4.get()}\n Lớp: {input5.get()}")  # Lấy dữ liệu người dùng nhập
   data11 = input.get() == "" or input2.get() == "" or input3.get() == "" or input4.get() == "" or input5.get() == ""
   if data11 is True:
       messagebox.showwarning("Cảnh báo", "Mời bạn nhập data hoặc link")
   else:
       img = qrcode.make(data)  # tạo qr code
       img.save(direct)  # lưu qr code dưới dạng ảnh theo đường dẫn đã chọn
       msb_result = messagebox.showinfo("Thông báo", "Bạn đã tạo xong mã QR")


imageFrame = Canvas(window, borderwidth= 0, height=0, width=600)
imageFrame.pack(anchor=W, expand=YES, fill=BOTH)
def open_qr():
   fileName = filedialog.askopenfilename(title='image need')
   image = Image.open(fileName)
   width, height = image.size
   size = f'{width} x {height}'




   imageSize = image.resize((300, 300), Image.ANTIALIAS)
   imageDisplay = ImageTk.PhotoImage(imageSize)
   imagetk = imageFrame.create_image(480, 20, image=imageDisplay, anchor=NW)
   labelFileName['text'] = fileName






# hàm tạo nút xóa dữ liệu
def clear():
   input.set('')
   entry_path.set('')
   input2.set('')
   input3.set('')
   input4.set('')
   input5.set('')




# tạo hộp nhập dữ liệu Họ và tên==========
link = Label(window, text="Họ và tên", fg='black', font='arial 10', width='10')
link.place(x=20, y=20)


input = StringVar()
input_data = Entry(window, width=55, textvariable=input)
# dòng này để định vị hộp nằm ở đâu
input_data.place(x=120, y=20)


# tạo hộp nhập dữ liệu Nơi lưu file==============
path = Label(window, text="Nơi lưu file", fg='black', font='arial 10', width='10')
path.place(x=20, y=170)


entry_path = StringVar()
link_path = Entry(window, width=35, textvariable=entry_path)
link_path.place(x=120, y=170)


# 2============== tự làm MSSV
# tạo label
data2 = Label(window, text="Ngày sinh", fg='black', font='arial 10', width='10')
data2.place(x=20, y=50)


# tạo hộp nhập dữ liệu
input2 = StringVar()
input2_data2 = Entry(window, width=35, textvariable=input2)
input2_data2.place(x=120, y=50)


# 3================== Năm sinh
data3 = Label(window, text="Mã số SV", fg='black', font='arial 10', width='10')
data3.place(x=20, y=80)


# tạo hộp nhập dữ liệu
input3 = StringVar()
input3_data3 = Entry(window, width=35, textvariable=input3)
input3_data3.place(x=120, y=80)


# 4========================Ngành học
data4 = Label(window, text="Ngành học", fg='black', font='arial 10', width='10')
data4.place(x=20, y=110)


# tạo hộp nhập dữ liệu
input4 = StringVar()
input4_data4 = Entry(window, width=35, textvariable=input4)
input4_data4.place(x=120, y=110)


# 5========================Lớp
data5 = Label(window, text="Lớp", fg='black', font='arial 10', width='10')
data5.place(x=20, y=140)


# tạo hộp nhập dữ liệu
input5 = StringVar()
input5_data5 = Entry(window, width=35, textvariable=input5)
input5_data5.place(x=120, y=140)


# Tạo button brower


btn_brower = Button(window, width=10, text="Duyệt", command=open_path)
btn_brower.place(x=368, y=170)


# Tạo button creat
# command=getqr để tạo mã QR
btn_creat = Button(window, width=10, text="Tạo QR Code", command=getqr)
btn_creat.place(x=240, y=200)


# Tạo Button mở mã QR
btn_open_qr = Button(window, width=10, text="Mở mã QR",  command= open_qr )
btn_open_qr.place(x=368, y=200 )


# tạo nút clear
clear1 = Button(window, width=10, text="Xóa tất cả", command=clear)
clear1.place(x=140, y=200)


window.mainloop()



```
